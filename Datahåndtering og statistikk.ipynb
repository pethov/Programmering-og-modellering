# Lesing av fil

Vi kan lese filer vha. funksjonen _loadtxt_:

data = loadtxt('filnavn',skiprows=1,delimiter=',',dtype=str,usecols=(1,2,4))

```python
from pylab import *

data = loadtxt('sunspots.csv', delimiter=',', skiprows=1)

aar = data[:,0]
solflekker = data[:,1]

plot(aar,solflekker,color='seagreen')
xlabel('Ã…r')
ylabel('Antall solflekker')
show()
```
